```python
def reformat_limits(restrictions):
    try:
        for i in range(len(restrictions)):
            if '<' in restrictions[i]:
                restrictions[i] = restrictions[i][:restrictions[i].index('<')].replace(' ', '')
            else:
                restrictions[i] = restrictions[i][:restrictions[i].index('=')].replace(' ', '')
            restrictions[i] = restrictions[i]
        return restrictions
    except ValueError:
        return restrictions
```
